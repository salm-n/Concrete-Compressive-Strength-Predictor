```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv("Concrete_Data_Yeh_final.csv")
data = data.dropna()
# Create a scatter plot for each column in the data against the y column
for col in data.columns[:-1]:
    try:
        # Filter out 0 x values
        filtered_data = data[data[col] != 0]

        plt.figure(figsize=(10, 6))
        plt.scatter(filtered_data[col], filtered_data["csMPa"])

        # Calculate the linear regression parameters
        m, b = np.polyfit(filtered_data[col], filtered_data["csMPa"], 1)

        # Plot the linear regression line
        plt.plot(filtered_data[col], m * filtered_data[col] + b, color='red')

        # Calculate the r-squared value
        r2 = np.corrcoef(filtered_data[col], filtered_data["csMPa"])[0, 1] ** 2

        # Display the r-squared value and equation
        plt.text(20, 80, f'r = {r2:.2f}', fontsize=12)
        plt.text(20, 60, f'y = {m:.2f}x + {b:.2f}', fontsize=12)

        plt.xlabel(col)
        plt.ylabel("csMPa")
        plt.title(f"Scatter plot of {col} vs. csMPa")
        plt.grid(True)
        plt.show()
    except:
        pass  # Ignore any errors
```
